```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/diabetes.csv')
```

Setting a seed ensures consistent random sampling, allowing others to
replicate the exact results.

```python
np.random.seed(42)   #Set a seed for reproducibility
```

```python
sample = data.sample(n=25)   #Random sample of 25 observations
```

---->   set a seed (to ensure work reproducibility) and take a random sample of 25 observations and
find the mean Glucose and highest Glucose values of this sample and compare these statistics
with the population statistics of the same variable. You should use charts for this comparison

```python
#Statistics for the sample
sample_mean_glucose = sample['Glucose'].mean()
sample_maximum_glucose = sample['Glucose'].max()
```

```python
# Population statistics
population_mean_glucose = data['Glucose'].mean()
population_maximum_glucose = data['Glucose'].max()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(['Sample Mean', 'Sample Max', 'Population Mean', 'Population Max'],
        [sample_mean_glucose, sample_maximum_glucose, population_mean_glucose, population_maximum_glucose])
plt.title('Comparison of Glucose Statistics (Sample vs Population)')
plt.xlabel('Statistic')
plt.ylabel('Glucose Value')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(['Sample Mean', 'Sample Max', 'Population Mean', 'Population Max'],
        [sample_mean_glucose, sample_maximum_glucose, population_mean_glucose, population_maximum_glucose], marker='o')
plt.title('Comparison of Glucose Statistics (Sample vs Population)')
plt.xlabel('Statistic')
plt.ylabel('Glucose Value')
plt.grid(True)
plt.show()
```

---->   Using bootstrap (replace= True), create 500 samples (of 150 observation each) from the
population and find the average mean, standard deviation and percentile for BloodPressure and
compare this with these statistics from the population for the same variable. Again, you should
create charts for this comparison

```python
num_bootstrap_samples = 500
sample_size = 150
```

```python
for i in range(num_bootstrap_samples):
    # Sample with replacement
    bootstrap_sample = data.sample(n=sample_size, replace=True)

    # Calculate mean, standard deviation, and percentile
    bootstrap_mean = bootstrap_sample['BloodPressure'].mean()
    bootstrap_std = bootstrap_sample['BloodPressure'].std()
    bootstrap_percentile = np.percentile(bootstrap_sample['BloodPressure'], 75)
```

```python
# Calculate average statistics for bootstrap samples
avg_bootstrap_mean = np.mean(bootstrap_mean)
avg_bootstrap_std = np.mean(bootstrap_std)
avg_bootstrap_percentile = np.mean(bootstrap_percentile)
```

```python
#Population statistics
population_mean_bp = data['BloodPressure'].mean()
population_std_bp = data['BloodPressure'].std()
population_percentile_bp = np.percentile(data['BloodPressure'], 75)
```

```python
plt.figure(figsize=(8, 5))
plt.subplot(1, 3, 1)
sns.barplot(x=['Population', 'Bootstrap'], y=[population_mean_bp, avg_bootstrap_mean])
plt.title('Comparison of Mean Blood Pressure')
plt.ylabel('Mean Blood Pressure')
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 2)
plt.plot(['Population', 'Bootstrap'], [population_mean_bp, avg_bootstrap_mean])
plt.title('Comparison of Mean Blood Pressure')
plt.ylabel('Mean Blood Pressure')
plt.tight_layout()
plt.show()
```

```python
plt.subplot(1, 3, 2)
sns.barplot(x=['Population', 'Bootstrap'], y=[population_std_bp, avg_bootstrap_std])
plt.title('Comparison of Standard Deviation of Blood Pressure')
plt.ylabel('Standard Deviation')


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))  # Adjust the figure size as needed
plt.plot(['Population', 'Bootstrap'], [population_std_bp, avg_bootstrap_std], marker='o')  # Line plot
plt.title('Comparison of Standard Deviation of Blood Pressure')  # Set the title of the plot
plt.xlabel('Group')  # Set the label for the x-axis
plt.ylabel('Standard Deviation')  # Set the label for the y-axis
plt.grid(True)  # Add grid lines
plt.show()
```

```python
plt.subplot(1, 3, 3)
sns.barplot(x=['Population', 'Bootstrap'], y=[population_percentile_bp, avg_bootstrap_percentile])
plt.title('Comparison of 75th Percentile of Blood Pressure')
plt.ylabel('75th Percentile')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(x=['Population', 'Bootstrap'], y=[population_percentile_bp, avg_bootstrap_percentile])
plt.title('Comparison of 75th Percentile of Blood Pressure')
plt.xlabel('Metric')
plt.ylabel('75th Percentile')
plt.tight_layout()

```
